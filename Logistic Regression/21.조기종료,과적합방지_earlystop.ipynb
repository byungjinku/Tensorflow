## 텐서플로우에서 조기종료

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = tf.placeholder(tf.float32) #예측값으로 활용하기 위해 변수로 준다
y = tf.constant([1,2,3], tf.float32) #실제값(label)이기 때문에 고정값으로 준다
```

```python
# w = tf.Variable(10.)
# b = tf.Variable(10.)
w = tf.get_variable('w1', shape=[1], initializer=tf.contrib.layers.xavier_initializer()) #사비어 초기화 알고리즘에 의해 ..
b = tf.get_variable('b1', shape=[1], initializer=tf.contrib.layers.xavier_initializer())
# xavier, he algorithm (초기화 알고리즘 - 가장 최적의 초기값을 부여해준다.)
#tf.contrib.layers.variance_scaling_initializer()
```

```python
hx = w*x + b #예측값
cost = tf.reduce_mean(tf.square(hx-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1) #Learning rate
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
hist_loss=[] #list선언
patience = 16 #16번 학습을 시킨다. #16번 기다리겠다
min_delta = 0.001 #엄격하게 주려면 값을 작게 주면 된다. #앞의 값 과의 차이

for i in range(300):
    sess.run(train,{x:[1,2,3]}) 
    c = sess.run(cost,{x:[1,2,3]})
    hist_loss.append(c)
    print(i,c)
    
    if i>0:
        if hist_loss[i-1]-hist_loss[i]>min_delta:    #이전값과 비교해준다. 
            pcnt=0   
        else:
            pcnt+=1
        
        if pcnt>patience:
            print('early stopping...........')
            break
#tensorflow는 patience와 min_delta를 직접 써주어야 한다. 
```

싸비어 초기화 전: 123번만에 조기종료/ 싸비어 초기화 후: 67번만에 조기종료

1. 반복문은 충분히 돌리면 된다
 
2. 이전데이터랑 비교해서 일정값 이상 변화가 없으면 변화가 없으면 break해서 빠져나간다 (반복문 종료)

## 케라스에서 조기종료

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
x = [1,2,3]
y = [1,2,3]
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
IO = Dense(units=1, input_dim=1)
model = Sequential([IO])
model.compile(loss = 'mean_squared_error',optimizer=Adam(0.1))
early = EarlyStopping(monitor='loss', min_delta=0.001, patience=16)
h = model.fit(x,y,epochs=1000, callbacks=[early])
```

28만에 조기종료 했다
