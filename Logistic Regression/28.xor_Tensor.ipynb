```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection
# from tensorflow.keras.backend import one_hot #session run이 필요하다. 
from tensorflow.keras.utils import to_categorical #session이 필요 없기 때문에 이걸 쓰면 된다. 굳이 one-hot으로 변경 안해도 된다.
import matplotlib.pyplot as plt
```

```python
xdata = np.array([[0,0],[0,1],[1,0],[1,1]],dtype=np.float32) #(4,2)
ydata = np.array([[0],[1],[1],[0]],dtype=np.float32) 
```

```python
x = tf.placeholder(tf.float32,[None,2])
y = tf.constant(ydata, tf.float32)
```

```python
w1 = tf.Variable(tf.random_uniform([2,2])) #두번 안하고 한번에 계산해 주기 위해
b1 = tf.Variable(tf.random_uniform([2]))
z1 = tf.matmul(x,w1) + b1 #(4,2)
lay1 = tf.nn.sigmoid(z1)  #결과적으로 y1,y2가 나온다.
```

```python
w2 = tf.Variable(tf.random_uniform([2,1])) 
b2 = tf.Variable(tf.random_uniform([1]))
z2 = tf.matmul(lay1,w2) + b2 
hx = tf.nn.sigmoid(z2)  
```

```python
# cost_i = y*(-tf.log(hx))+(1-y)*(-tf.log(1-hx))
cost_i=tf.nn.sigmoid_cross_entropy_with_logits(logits=z2,
                                            labels=y)
cost = tf.reduce_mean( cost_i )
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1) #learning rate
train = optimizer.minimize(cost)
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(30000):
    sess.run(train, {x:xdata}) 
    c = sess.run(cost,{x:xdata})
    print(i,c)
```

```python
h = sess.run(hx,{x:xdata})
t=tf.cast(h>0.5,tf.float32)
sess.run(t)
```
