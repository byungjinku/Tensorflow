```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/data-01.csv',header=None)
df.columns=['q1','q2','mid','score']
df
```

```python
xdata = df[['q1','q2','mid']]
ydata = df[['score']]
```

```python
x = tf.placeholder( tf.float32, shape=[None,3])
y = tf.constant( ydata ,tf.float32)
```

```python
# [3,1] #[ 피쳐갯수, 라벨갯수]
w = tf.Variable( tf.random_uniform([3,1]) )
b = tf.Variable( tf.random_uniform([1]) )
```

```python
hx = tf.matmul(x,w) +b # nX3 3X1
cost = tf.reduce_mean(tf.square(hx-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.000001)#learning rate
train = optimizer.minimize( cost )
```

```python
sess = tf.Session()
init =tf.global_variables_initializer()
sess.run( init)
```

```python
for i in range(10000):
    sess.run(train, {x:xdata})
    print(i, sess.run(cost, {x:xdata}))
```
