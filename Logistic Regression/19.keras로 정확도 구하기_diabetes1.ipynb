```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/diabetes1.csv', header=None) #칼럼명 없음. 데이터 미뤄주기 위해. 
df
```

```python
xdata = df.iloc[:,:-1]
ydata = df.iloc[:,[-1]] #벡터가 아니라 행렬로 추출하기 위해. 맨 마지막 값이 label 이다. 
```

```python
IO = Dense(units=1, input_dim=8, activation='sigmoid') #sigmoid를 활성함수로 쓰겠다. 
model = Sequential( [IO] ) #순차모델
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),metrics=['accuracy'])  #loss는 로그 함수를 사용한다. 
h = model.fit( xdata, ydata, epochs=1000)
```

```python
model.predict(xdata)
```

```python
model.predict_classes(xdata) #케라스 버전에서는 얘를 사용한다. 
```

```python
h.history['acc'][-1] #맨 마지막꺼 가져왔다. 
```
