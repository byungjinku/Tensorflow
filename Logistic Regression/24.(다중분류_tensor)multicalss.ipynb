```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn import datasets
from tensorflow.keras.backend import one_hot
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
iris
```

```python
iris['target']
```

```python
sess = tf.Session()
```

```python
t= one_hot(iris['target'],3)
sess.run(t)
```

```python
arr = iris['data'] #sl 5.1, sw 3.5, pl 1.4, pw 0.2 종분류 및 정확도
arr
```

```python
xdata = arr
ydata = t
```

```python
x = tf.placeholder(tf.float32,[None,4])
y = ydata

w = tf.Variable(tf.random_uniform([4,3]))
b = tf.Variable(tf.random_uniform([3]))

z = tf.matmul(x,w) + b
hx = tf.nn.softmax(z)
cost_i = tf.nn.softmax_cross_entropy_with_logits_v2(logits= z, labels=y) #logits 인자에는 hx함수 , labels= Tensor형 yvalue
cost = tf.reduce_mean(cost_i)

optimizer = tf.train.GradientDescentOptimizer(0.1) #learning rate
train = optimizer.minimize(cost)
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
hist_loss =[]
patience = 16
min_delta = 0.001 #이전값과 16번 비교하여 min_delta보다 작으면 빠져나가겠다. 
pcnt = 0
```

```python
for i in range(30000):
    sess.run(train,{x:xdata}) #벼림 연산 # w = w.assign(w-0.1*g)
    c = sess.run(cost,{x:xdata}) 
    hist_loss.append( c )
    print(i, c)
    if i > 0 :
        if hist_loss[i-1]-hist_loss[i] > min_delta:
            pcnt = 0
        else:
            pcnt+= 1
        if pcnt > patience:
            print ('early stopping...')
            break
```

# 확률값

```python
h = sess.run(hx,{x:[[5.1,3.5,1.4,0.2]]}) #hx값을 구하려면 z값을 구해야하고, z값을 구하려면 x값이 필요하기 때문에 x를 넣어준다. 
h
```

# 정확도

```python
accuracy = tf.reduce_mean(tf.cast(tf.equal(h2, iris['target']), dtype=tf.float32))
```

```python
sess.run(accuracy,{x:xdata})
```
