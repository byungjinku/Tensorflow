```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
import math
```

```python
df = pd.read_csv('data/diabetes1.csv', header=None) #칼럼명 없음. 데이터 미뤄주기 위해. 
df
```

```python
xdata = df.iloc[:,:-1]
ydata = df.iloc[:,[-1]] #벡터가 아니라 행렬로 추출하기 위해. 맨 마지막 값이 label 이다. 
```

```python
ydata
```

```python
xdata.shape #feature가 8개 이다. 
```

```python
x = tf.placeholder(tf.float32,shape=[None,8])
y = tf.constant(ydata, tf.float32)
```

```python
w = tf.Variable(tf.random_uniform([8,1])) #feature의 개수가 8개 이므로 wait도 8개가 나온다. w1*x1 + w2*x2 +...... w8*x8
b = tf.Variable(tf.random_uniform([1]))
```

```python
z = tf.matmul(x,w) + b
hx = tf.nn.sigmoid(z) #0 ~ 1사이의 값
```

```python
cost_i = y*(-tf.log(hx))+(1-y)*(-tf.log(1-hx))
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1) #learning rate
train = optimizer.minimize(cost)

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
predict = tf.cast(hx>0.5, dtype=tf.float32) 
accuracy = tf.reduce_mean(tf.cast(tf.equal(predict,y), dtype=tf.float32)) #equal(predict,y) = e
```

```python
for i in range(10000):
    sess.run(train, {x:xdata}) #x가 placeholder 이기 때문에 값을 지정해줘야 한다,
    print(i, sess.run(cost, {x:xdata}))
```

```python
sess.run(predict, {x:xdata})
```

```python
sess.run(accuracy,{x:xdata} )
```
