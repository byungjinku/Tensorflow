```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets #전처리...
import matplotlib.pyplot as plt
```

```python
data = datasets.load_diabetes()
data
```

```python
df = pd.read_csv('data/diabetes1.csv', header=None)
df
```

```python
xdata = df.iloc[:,:-1]
ydata = df.iloc[:,[-1]]
```

```python
ydata
```

```python
xdata.shape
```

```python
x = tf.placeholder(tf.float32,shape=[None,8] )
y = tf.constant( ydata, tf.float32)
```

```python
w = tf.Variable( tf.random_uniform([8,1]) )
b = tf.Variable( tf.random_uniform([1]) )
```

```python
z = tf.matmul(x,w) + b
hx = tf.nn.sigmoid( z ) # 0~ 1사이값
```

```python
cost_i = y*(-tf.log(hx))+(1-y)*(-tf.log(1-hx))
cost = tf.reduce_mean( cost_i )
```

```python
optimizer = tf.train.GradientDescentOptimizer( 0.1 )
train = optimizer.minimize( cost )
sess = tf.Session()
init =tf.global_variables_initializer()
sess.run( init)
```

```python
predict = tf.cast( hx>0.5, dtype=tf.float32)
accuracy = tf.reduce_mean( tf.cast(tf.equal(predict,y), dtype=tf.float32) )
```

```python
for i in range(1000):
    sess.run( train, {x:xdata})
    print( i, sess.run(cost, {x:xdata}))
```
