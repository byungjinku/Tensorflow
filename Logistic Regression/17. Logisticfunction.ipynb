```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
xdata = [[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]] #공부한 시간, 출석일수
ydata = [[0],[0],[0],[1],[1],[1]] #0:불합격, 1:합격
```

```python
x = tf.placeholder(tf.float32,shape=[None,2])
y = tf.constant(ydata, tf.float32)
```

```python
w = tf.Variable(tf.random_uniform([2,1]))
b = tf.Variable(tf.random_uniform([1]))
```

```python
z = tf.matmul(x,w) + b
hx = tf.nn.sigmoid(z) 
```

```python
cost_i = y*(-tf.log(hx))+(1-y)*(-tf.log(1-hx))
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.AdamOptimizer(0.01) #learning rate
train = optimizer.minimize(cost)

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(1000):
    sess.run(train, {x:xdata})
    print(i, sess.run(cost, {x:xdata}))
```

```python
predict = tf.cast(h>0.5, dtype=tf.float32) 
e = tf.equal(predict,y) #예측값과 실제 값이 모두 같은가? predict == y
m = tf.reduce_mean(tf.cast(e, dtype=tf.float32))
#sess.run(e) 
sess.run(m)
```
