```python
data = iris['data']
```

```python
data.shape
```

```python
ydata= data[:,[0]]
xdata= data[:,1:]
```

```python
xdata
```

```python
x = tf.placeholder(tf.float32,shape=[None,3])
y = tf.constant(ydata, tf.float32)
```

```python
w = tf.Variable(tf.random_uniform([3,1]))
b = tf.Variable(tf.random_uniform([1]))
```

```python
hx = tf.matmul(x,w) + b
cost = tf.reduce_mean(tf.square(hx-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.01)#learning rate
train = optimizer.minimize( cost )
```

```python
sess = tf.Session()
init =tf.global_variables_initializer()
sess.run( init)
```

```python
for i in range(10000):
    sess.run(train,{x:xdata})
    print(i, sess.run(cost, {x:xdata}))
```

```python
sess.run(w)
```

```python
sess.run(b)
```

```python
# 3.5, 1.4, 0.2
sess.run(hx, {x:[[3.5,1.4,0.2]]})
```

```python
plt.plot(ydata)
plt.plot(sess.run(hx,{x:xdata}), 'r--')
plt.show()
```
