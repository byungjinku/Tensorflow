```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
xdata = np.array( [[0,0],[0,1],[1,0],[1,1]],dtype=np.float32 )
ydata = np.array( [[0],[1],[1],[0]],dtype=np.float32 )
```

```python
x = tf.placeholder( tf.float32,[None,2] )
y = tf.constant( ydata, tf.float32 )
```

```python
w1 = tf.Variable(tf.random_uniform([2,2]))
b1 = tf.Variable( tf.random_uniform([2]) )
z1 = tf.matmul(x,w1) +b1
lay1 = tf.nn.sigmoid( z1 ) # 4X2  y1 y2
```

#### 이 과정이 중요함

```python
w2 = tf.Variable(tf.random_uniform([2,1]))
b2 = tf.Variable( tf.random_uniform([1]) )
z2 = tf.matmul(lay1,w2) +b2
hx = tf.nn.sigmoid( z2 )
```

```python
cost_i=tf.nn.sigmoid_cross_entropy_with_logits(logits=z2,
                                            labels=y)
cost = tf.reduce_mean( cost_i )
```

```python
optimizer = tf.train.GradientDescentOptimizer( 0.1 )
train = optimizer.minimize( cost )
sess = tf.Session()
init =tf.global_variables_initializer()
sess.run( init)
```

```python
for i in range(3000):
    sess.run(train,{x:xdata}) #벼림 연산 # w = w.assign(w-0.1*g)
    c = sess.run(cost,{x:xdata }) 
    print(i,c)
```
