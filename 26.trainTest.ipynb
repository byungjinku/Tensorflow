```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
xdata = iris['data']
```

```python
ydata =to_categorical( iris['target'],3)
```

```python
a,b,c,d = [np.array([1,2]), np.array([3,4]),
           np.array([5,6]),np.array([7,8])]
```

```python
x_train,x_test, y_train,y_test =model_selection.train_test_split( xdata,ydata, 
                                                                 test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
x = tf.placeholder(tf.float32,[None,4])
y = y_train

w = tf.Variable(tf.random_uniform([4,3]))
b = tf.Variable(tf.random_uniform([3]))
```

```python
z = tf.matmul(x,w) + b
hx = tf.nn.softmax(z)
cost_i = tf.nn.softmax_cross_entropy_with_logits_v2(logits= z, labels=y) #logits 인자에는 hx함수 , labels= Tensor형 yvalue
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1) #learning rate
train = optimizer.minimize(cost)
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(3000):
    sess.run(train,{x:x_train}) #벼림 연산 # w = w.assign(w-0.1*g)
    c = sess.run(cost,{x:x_train}) 
    print(i, c)
```
