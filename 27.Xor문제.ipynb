```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
xdata = ([[0,0], [0,1], [1,0], [1,1]], dtype=np.float32)
ydata = ([[0],[1],[1],[0]], dtype=np.float32)
```

```python
x = tf.placeholder(tf.float32,[None,2])
y = tf.constant(ydata, tf.float32)
```

```python
w = tf.Variable(tf.random_uniform([2,1]))
b = tf.Variable(tf.random_uniform([1]))
```

```python
z = tf.matmul(x,w) + b
hx = tf.nn.sigmoid( z ) # 0~ 1사이값
```

```python
cost_i =tf.nn.sigmoid_cross_entropy_with_logits(logits=z, labels=y)
cost = tf.reduce_mean( cost_i )
```

```python
optimizer = tf.train.GradientDescentOptimizer( 0.1 )
train = optimizer.minimize( cost )
sess = tf.Session()
init =tf.global_variables_initializer()
sess.run( init)
```

```python
for i in range(1000):
    sess.run( train, {x:xdata})
    c = sess.run(cost,{x:xdata})
    print( i, c)
```

```python
h = sess.run(hx, {x:xdata})
t = tf.cast(h>0.5, tf.float32)
sess.run(t)
```
